```python
import pandas as pa
import numpy as np
import matplotlib.pyplot as plt
```

```python
ta = pa.read_csv('/home/wooseok/WISE_data/3rdnewwise/3rd_obsb_outcut_md_taurus_NEOWISE_avg2_median_maxmin_4000.csv')
ta
```

```python
ta = ta[(ta.N_w2 > 5) &
         (ta.dist_sd < 0.3) &
         (ta.avg_eW2 < 0.2) ]

ta
```

```python
#P,D,PMS+E 로 설정

ta.loc[(ta['class'] == "P") |
        (ta['class'] == "F") |
        (ta['class'] == "FP") |
        (ta['class'] == "0") |
        (ta['class'] == "I") |
        (ta['class'] == "I?"),'class'] = 'P'


ta.loc[(ta['class'] == "D") |
        (ta['class'] == "II") |
        (ta['class'] == "full") |
        (ta['class'] == "full?") |
        (ta['class'] == "debris/ev trans") |
        (ta['class'] == "transitional") |
        (ta['class'] == "evolved") |
        (ta['class'] == "ev or trans")
        ,'class'] = 'D'

ta.loc[(ta['class'] == "E") |
        (ta['class'] == "III") 
        ,'class'] = 'PMS+E'

tar = ta.loc[(ta['class'] == "PMS+E") | 
      (ta['class'] == "D") |
      (ta['class'] == "P"),:]

tar['class'].value_counts()
```
